SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
# Определим последовательность шагов воронки
funnel_steps = [
    'install / open_web',
    'introduction',
    'registration',
    'main_page',
    'onboarding_complete',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# Фильтруем датасет, оставляя только шаги воронки
funnel_df_steps = df[df['event_name'].isin(funnel_steps)]

# Считаем число уникальных пользователей на каждом шаге
funnel_counts = funnel_df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)

# Собираем таблицу воронки
funnel_df = pd.DataFrame({
   'step': funnel_steps,
   'users': funnel_counts.values
})

# Добавляем конверсии для классической воронки и для ste-by-step воронки
funnel_df['conversion_from_first_%'] = (funnel_df['users'] / funnel_df['users'].iloc[0] * 100).round(1)
funnel_df['conversion_from_previous_%'] = (funnel_df['users'] / funnel_df['users'].shift(1) * 100).round(1)

display(funnel_df.head(8))
````

````python
# Построим классическую воронку
plt.figure(figsize=(16, 6))

bars = plt.bar(funnel_df['step'], funnel_df['conversion_from_first_%'], color='skyblue')

# Добавляем подписи в процентах сверху каждого столбца
for bar, value in zip(bars, funnel_df['conversion_from_first_%']):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  
        height,                             
        f'{value:.1f}%',
        ha='center', va='bottom'
    )
    
plt.title('Классическая воронка')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()
````

````python
# Построим step-by-step воронку
plt.figure(figsize=(16, 6))

funnel_df_filtered = funnel_df[funnel_df['step']!='install / open_web']
bars = plt.bar(funnel_df_filtered['step'], funnel_df_filtered['conversion_from_previous_%'], color='skyblue')

# Добавляем подписи в процентах сверху каждого столбца
for bar, value in zip(bars, funnel_df_filtered['conversion_from_previous_%']):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  
        height,                             
        f'{value:.1f}%',
        ha='center', va='bottom'
    )

plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.show()
````

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.  

````python
def get_risk_df(df):

    # Отфильтруем пользователей, которые вложились в первый депозит asset_dffirst_deposit)
    first_deposit_df = df[df['event_name'] == 'first_deposit']

    # Определим дату первого депозита для каждого пользователя
    first_deposit_ts_df = first_deposit_df.groupby('user_id')['event_ts'].min().reset_index(name='first_deposit_ts')

    # Отфильтруем пользователей, которые купили активы (asset_df = asset_purchase)
    asset_df = df[(df['event_name'] == 'asset_purchase')]

    # Объеденим и выберем только те активы, которые были купдены пользователем после первого депозита
    asset_first_deposit_merged = pd.merge(asset_df, first_deposit_ts_df, on = ['user_id'], how = 'inner').query('event_ts > first_deposit_ts')

    # Определим дату первой покупки актива
    asset_event_ts_df = asset_first_deposit_merged.groupby('user_id')['event_ts'].min().reset_index(name='first_asset_ts')

    # Отфильтруем пользователей, которые вложились во второй депозит (asset_df = second_deposit)
    second_deposit_df = df[df['event_name']=='second_deposit']


    # Добавим флаг наличия второго депозита
    asset_first_deposit_merged['second_deposit_flag'] = asset_first_deposit_merged['user_id'].isin(second_deposit_df['user_id'])
    
    # Сгруппируем по уровню риска и посчитаем конверсию во второй депозит
    risk_df = asset_first_deposit_merged.groupby('risk_level').agg(
        user_count = ('user_id', 'nunique'),
        second_deposit_user_count = ('second_deposit_flag', 'sum')
    ).reset_index()

    risk_df['risk_conversion'] = (100*risk_df['second_deposit_user_count']/risk_df['user_count']).round(2)
    
    return risk_df

risk_df = get_risk_df(df)

display(risk_df.head())
````

````python
risk_df = risk_df.sort_values('risk_conversion')

# Построим столбчатую диаграмму конверсии
plt.figure(figsize=(16, 6))

bars = plt.bar(risk_df['risk_level'], risk_df['risk_conversion'], color='skyblue')

# Добавляем подписи в процентах сверху каждого столбца
for bar, value in zip(bars, risk_df['risk_conversion']):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  
        height,                             
        f'{value:.2f}%',
        ha='center', va='bottom'
    )

plt.title('Диаграмма конверсии во второй депозит с разбивкой по риску актива')
plt.ylabel('Конверсия во второй депозит(%)')
plt.xlabel('Уровень риска')
plt.show()
````

**Промежуточный вывод:**

В датасете представлены три уровня риска активов:
- `Hight` - высокий,
- `Medium` - средний,
- `Low` - низкий.

По графику видим, что уровень риска актива достаточно сильно влияет на открытие второго депозите. Пользователи, купившие высокорисковые активы, только в 25.21% открывают второй депозит. При этом поведение пользователей со среднерисковыми и низкорисковыми активами практически одинаковое:

- пользователи с medium активами открывают второй депозит в 41.07% случаев
- пользователи с low активами открывают второй депозит в 44.32% случаев

- Проанализируйте, как категория риска купленного актива влияет на вероятность открытия второго депозита в тестовой и контрольной группах. 
- Сравните результаты и оцените, есть ли статистически значимая разница между группами. 

````python
# Для Control группы
# Отфильтруем датасет abt_df по группам
control_df = abt_df[abt_df['group'] == 'control']

# Подготовим risk-датасеты по группам
risk_control_df = get_risk_df(control_df)

print('Группа Control:')
display(risk_control_df)
````

````python
# Для Test группы
# Отфильтруем датасет abt_df по группам
test_df = abt_df[abt_df['group'] == 'test']

# Подготовим risk-датасеты по группам
risk_test_df = get_risk_df(test_df)

print('Группа Test:')
display(risk_test_df)
````

Для каждого уровня риска сформулируем гипотезу:
- H0: Доля пользователей, вложившихся во второй депозит, после покупки актива определенного риска, не изменилась — то есть разница в долях между группами статистически НЕ значима.
- H1: Доля пользователей, вложившихся во второй депозит, после покупки актива определенного риска, статистически значимо УВЕЛИЧИЛАСЬ в тестовой группе по сравнению с контрольной.

Проверим правостороннюю гипотезу, так как нас больше интересует увеличение прибыли, а соответственно, и увеличение конверсии в покупку второго депозита. 

Уровень значимости alpha возьмем равное 0.05

````python
# Выполним статистическую проверку с помощью Z-test пропорций
alpha = 0.05

stat_results = []

for risk_level in risk_control_df['risk_level'].unique():
    data_control = risk_control_df[risk_control_df['risk_level'] == risk_level].iloc[0]
    data_test = risk_test_df[risk_test_df['risk_level'] == risk_level].iloc[0]
    
    count = [data_test['second_deposit_user_count'], data_control['second_deposit_user_count']]
    
    nobs = [data_test['user_count'], data_control['user_count']]

    _, p_val = proportions_ztest(count, nobs, alternative='larger')

    stat_results.append({
        'risk_level': risk_level,
        'risk_conversion_control': data_control['risk_conversion'],
        'risk_convertion_test': data_test['risk_conversion'],
        'p_value': p_val.round(4)
    })
    
    if (p_val >= alpha):
         print(f'Разница между долями пользователей, вложившихся во второй депозит, после покупки {risk_level}-актива статистически НЕ значима.')
    else:
         print(f'Разница между долями пользователей, вложившихся во второй депозит, после покупки {risk_level}-актива статистически значима.')


stat_results = pd.DataFrame(stat_results)
display(stat_results)
````

**Промежуточный вывод:**

При проверке правосторонной гипотезы:

- H0: Доля пользователей, вложившихся во второй депозит, после покупки актива определенного риска, не изменилась — то есть разница в долях между группами статистически НЕ значима.
- H1: Доля пользователей, вложившихся во второй депозит, после покупки актива определенного риска, статистически значимо УВЕЛИЧИЛАСЬ в тестовой группе по сравнению с контрольной.

выяснили, что уровень риска актива влияет на долю пользователей, вложившихся во второй дипозит. Пользователи, купившие активы с уровнем риска high и low статистически значимо чаще открывают второй дипозит. .

Проверим гипотезу:

- H0: Средняя сумма депозитов на одного пользователя в контрольной и тестовой группах равна.
- H1: Средняя сумма депозитов на одного пользователя в контрольной и тестовой группах НЕ равна.

Уровень статистической значимости возьмем 0.05

Проверим гипотизу в обе стороны, так как средняя сумма всех депозитов на одного пользователя в группа оказалась практически равной. Будем использовать тест Уэлча, так как его можно применять к датасетам с выбросами. 

````python
# Проверим статистическую значимость.
alpha = 0.05

# Проведем t-тест для независимых выборок (Welch)
stat_avg_deposit_per_user, p_key_avg_deposit_per_user = ttest_ind(
    deposit_merged_control['total_amount_per_user'], 
    deposit_merged_test['total_amount_per_user'], 
    equal_var=False
)

if p_key_avg_deposit_per_user >= alpha:
    print('Различия между средними суммами всех депозитов на одного пользователя в контрольной и тестовой группах статистически НЕ значимы.')
else:
    print('Различия между средними суммами всех депозитов на одного пользователя в контрольной и тестовой группах статистически значимы.')

print(f'p_value = {p_key_avg_deposit_per_user.round(4)}')
````

Посмотрим накопленную динамику изменения средней суммы всех депозитов на одного пользователя по дням эксперимента для каждой группы

````python
# Рассмотрим накопленную динамику изменения по дням эксперимента для каждой группы
alpha = 0.05
def get_amount_df(df):
    
    results = []
    df = df.copy()
    df['day_active'] = df['event_ts'].dt.date
    for date in sorted(df['day_active'].unique()):
        df_temp = df[
            (df['event_ts'] >= pd.Timestamp('2025-06-02 00:00:00')) & 
            (df['event_ts'] < pd.Timestamp(date) + pd.Timedelta(days=1))]
    
        df_temp_control = df_temp[df_temp['group'] == 'control']
        
        total_deposit_control = (
            df_temp_control[df_temp_control['event_name'].isin(['first_deposit', 'second_deposit'])]
            .groupby('user_id')['amount']
            .sum()
            .reindex(df_temp_control['user_id'].unique(), fill_value=0)
        )
        
        df_temp_test = df_temp[df_temp['group'] == 'test']
        
        total_deposit_test = (
            df_temp_test[df_temp_test['event_name'].isin(['first_deposit', 'second_deposit'])]
            .groupby('user_id')['amount']
            .sum()
            .reindex(df_temp_test['user_id'].unique(), fill_value=0)
        )
        
    
        _, p_value = ttest_ind(
            total_deposit_control, 
            total_deposit_test, 
            equal_var=False
        )
    
        # Создаем DataFrame для текущей даты
        df_row = pd.DataFrame({
            'date': [date],
            'p_value': [p_value],
            'avg_control': [total_deposit_control.mean()],
            'avg_test': [total_deposit_test.mean()]
        })
    
        results.append(df_row)

    # Объединяем все результаты в один DataFrame
    df_data = pd.concat(results, ignore_index=True)

    # Сортируем по дате
    df_data.sort_values('date', inplace=True)
    return df_data

data = get_amount_df(abt_df)

plt.figure(figsize = (12,6))

plt.plot(data['date'], data['avg_control'], marker = 'o', label = 'Control group')
plt.plot(data['date'], data['avg_test'], marker = 'o', label = 'Test group')

plt.title('Накопленная динамики изменения средней суммы депозита на одного пользователя по дням эксперимента для каждой группы')
plt.xlabel('Дата')
plt.ylabel('Средняя сумма депозита на одного пользователя')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize = (12,6))

plt.plot(data['date'], data['p_value'], marker = 'o')

plt.axhline(y=alpha, color='r', linestyle='--', label='alpha = 0.05')

plt.title('Cтабильность p-value во время эксперимента')
plt.xlabel('Дата')
plt.ylabel('p_value')
plt.grid(True)
plt.show()
````

**Промежуточный вывод:**

Несмотря на то, что средняя сумма всех депозитов на одного пользователя в тестовой группе (92.46) оказалась немного выше, чем в контрольной (90.24), статистический тест показал, что различия между средними суммами всех депозитов на одного пользователя в контрольной и тестовой группах статистически НЕ значимы. Эти выводы подверждают и графики:
- на первом графике средния сумма всех депозитов тестовой группы первые 8 дней немного отставала от контрольной, а далее значения сравнялись и тестовая группа стала показывать результат лучше, чем контрольная.

**2. Барьерная метрика — конверсия из регистрации в первый депозит**

````python
def get_reg_to_first_deposit_df(df):
    
    # Отфильтруем данные по типу события
    registrations = df[df['event_name'] == 'registration']['user_id'].drop_duplicates()
    deposits_first = df[df['event_name'] == 'first_deposit']['user_id'].drop_duplicates()

    # Создаем DataFrame из регистраций
    result = pd.DataFrame({'user_id': registrations})

    # Проверяем наличие пользователя в депозитах и создаем колонку
    result['first_deposit_flag'] = result['user_id'].isin(deposits_first).astype(int)

    return result

def get_convertion_reg_to_first_deposit(df):
    data = get_reg_to_first_deposit_df(df)
    result = data['first_deposit_flag'].sum()/data['user_id'].nunique()
    
    return result

convertion_reg_to_first_deposit_control = get_convertion_reg_to_first_deposit(control_df)
convertion_reg_to_first_deposit_test = get_convertion_reg_to_first_deposit(test_df)

abs_convertion_reg_to_first_deposit = convertion_reg_to_first_deposit_test - convertion_reg_to_first_deposit_control
rel_convertion_reg_to_first_deposit = (convertion_reg_to_first_deposit_test - convertion_reg_to_first_deposit_control)/convertion_reg_to_first_deposit_control

print(f"Конверсия из registration в first_deposit в контрольной группе: {convertion_reg_to_first_deposit_control:.2%}")
print(f"Конверсия из registration в first_deposit в тестовой группе: {convertion_reg_to_first_deposit_test:.2%}")
print(f'Абсолютное значение (тестовая к контрольной группе) - {abs_convertion_reg_to_first_deposit:.2}')
print(f'Относительное значение (тестовая к контрольной группе) - {rel_convertion_reg_to_first_deposit:.2}')
````

Проверим статистическую значимость барьерной метрики:

Гипотеза:
- H0: Конверсия из регистрации в первый депозит равна в контрольной и тестовой группах
- H1: Конверсия из регистрации в первый депозит в тестовой группе меньше, чем в контрольной

Уровень значимости возьмем 0.05.

Метрика является конверсией, соответственно, будем использовать z-test с параметром smaller, то есть будем проверять, что метрика в тестовой группе НЕ уменьшилась.

````python
# Выполним статистическую проверку с помощью Z-test пропорций
alpha = 0.05

data_control = get_reg_to_first_deposit_df(control_df)
data_test = get_reg_to_first_deposit_df(test_df)

stat_results = []

count = [data_test['first_deposit_flag'].sum(), data_control['first_deposit_flag'].sum()]
    
nobs = [data_test['user_id'].nunique(), data_control['user_id'].nunique()]

_, p_val = proportions_ztest(count, nobs, alternative='smaller')
    
if (p_val >= alpha):
    print(f'Разница между конверсиями из регистрации в первый депозит в контрольной и тестовой группах статистически НЕ значима.')
else:
    print(f'Разница между конверсиями из регистрации в первый депозит в контрольной и тестовой группах статистически значима.')

print(f'p_value={p_val:.4}')
````

**Промежуточный вывод:**

Несмотря на то, что барьерная метрика `Конверсия из регистрации в первый депозит` тестовой группы (70.04%) показала меньший результат, чем в контрольной (71.54%), статистический тест с параметром  alternative='smaller' показал, что разница между этими метриками статистически НЕ значима. Это значит, что барьерная метрика в тестовой группе не ухудшается.

**3. Вспомогательная метрика 1 — конверсия из первого депозита во второй**

Метрика конверсия из первого депозита во второй не должна упасть, соответственно, метрика тестовой группы должна быть больше метрики в контрольной группе.

````python
def get_first_to_second_deposit_df(df):
    
    # Отфильтруем данные по типу события
    first_deposit_df = df[df['event_name'] == 'first_deposit']['user_id'].drop_duplicates()
    second_deposit_df = df[df['event_name'] == 'second_deposit']['user_id'].drop_duplicates()

    # Создаем DataFrame из регистраций
    result = pd.DataFrame({'user_id': first_deposit_df})

    # Проверяем наличие пользователя в депозитах и создаем колонку
    result['second_deposit_flag'] = result['user_id'].isin(second_deposit_df).astype(int)

    return result

def get_convertion_first_to_second_deposit(df):
    data = get_first_to_second_deposit_df(df)
    result = data['second_deposit_flag'].sum()/data['user_id'].nunique()
    
    return result

convertion_first_to_second_deposit_control = get_convertion_first_to_second_deposit(control_df)
convertion_first_to_second_deposit_test = get_convertion_first_to_second_deposit(test_df)

abs_convertion_first_to_second_deposit = convertion_first_to_second_deposit_test - convertion_first_to_second_deposit_control
rel_convertion_first_to_second_deposit = (convertion_first_to_second_deposit_test - convertion_first_to_second_deposit_control)/convertion_first_to_second_deposit_control

print(f"Конверсия из first_deposit в second_deposit контрольной группе: {convertion_first_to_second_deposit_control:.2%}")
print(f"Конверсия из first_deposit в second_deposit в тестовой группе: {convertion_first_to_second_deposit_test:.2%}")
print(f'Абсолютное значение (тестовая к контрольной группе) - {abs_convertion_first_to_second_deposit:.2}')
print(f'Относительное значение (тестовая к контрольной группе) - {rel_convertion_first_to_second_deposit:.2}')
````

Сформулируем гипотезу:

- H0: Разница между конверсиями из первого депозита во второй в контрольной и тестовой группах статистически НЕ значима.
- H1: Разница между конверсиями из первого депозита во второй в тестовой группе статистически значимо больше, чем в контрольной.

Уровень значимости будем использовать 0.05.

Метрика является конверсией, соответственно, будем использовать z-test с параметром larger, то есть будем проверять, что метрика в тестовой группе увеличилась.

````python
# Выполним статистическую проверку с помощью Z-test пропорций
alpha = 0.05

data_control = get_first_to_second_deposit_df(control_df)
data_test = get_first_to_second_deposit_df(test_df)

stat_results = []

count = [data_test['second_deposit_flag'].sum(), data_control['second_deposit_flag'].sum()]
    
nobs = [data_test['user_id'].nunique(), data_control['user_id'].nunique()]

_, p_val = proportions_ztest(count, nobs, alternative='larger')
    
if (p_val >= alpha):
    print(f'Разница между конверсиями из первого депозита во второй в контрольной и тестовой группах статистически НЕ значима.')
else:
    print(f'Разница между конверсиями из первого депозита во второй в тестовой группе статистически значимо больше, чем в контрольной.')

print(f'p_value={p_val:.4}')
````

Вспомогательная метрика `Конверсия из первого депозита во второй` должна подтвердить, что количество вторых депозитов увеличилось после обучения.

**Промежуточный вывод:**

Конверсия из первого депозита во второй в тестовой группе показала значение 32.61%, что выше, чем контрольной группе 20.26%. Для подтверждения данного результата был выполнен статистический тест, который подтвердил, что разница между конверсиями из первого депозита во второй в тестовой группе статистически значимо больше, чем в контрольной.

**3. Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит**

````python
def get_avg_deposit_per_paying_user_df(df):
    
    # Отфильтруем данные по типу события
    deposit_df = df[df['event_name'].isin(['first_deposit','second_deposit'])]['user_id'].drop_duplicates()
    result = df[df['user_id'].isin(deposit_df)].groupby('user_id')['amount'].sum().reset_index(name='total_amount')
    
    return result

def get_avg_deposit_per_paying_user(df):
    data = get_avg_deposit_per_paying_user_df(df)
    result = data['total_amount'].sum()/data['user_id'].nunique()
    
    return result

avg_deposit_per_paying_user_control = get_avg_deposit_per_paying_user(control_df)
avg_deposit_per_paying_user_test = get_avg_deposit_per_paying_user(test_df)

abs_avg_deposit_per_paying_user = avg_deposit_per_paying_user_test - avg_deposit_per_paying_user_control
rel_avg_deposit_per_paying_user = (avg_deposit_per_paying_user_test - avg_deposit_per_paying_user_control)/avg_deposit_per_paying_user_control

print(f"Средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит, в контрольной группе: {avg_deposit_per_paying_user_control:.5}")
print(f"Средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит, в тестовой группе: {avg_deposit_per_paying_user_test:.5}")
print(f'Абсолютное значение (тестовая к контрольной группе) - {abs_avg_deposit_per_paying_user:.2}')
print(f'Относительное значение (тестовая к контрольной группе) - {rel_avg_deposit_per_paying_user:.2}')
````

- H0: Разница между средними суммами всех депозитов на пользователя, который открыл хотя бы один депозит, статистически НЕ значима.
- H1: Разница между средними суммами всех депозитов на пользователя, который открыл хотя бы один депозит, в тестовой группе статистически значимо больше, чем в контрольной.

Уровень значимости будем использовать 0.05.

Метрика является средним значением, будем использовать t-test Уэлча с параметром larger, то есть будем проверять, что метрика в тестовой группе увеличилась.

````python
# Проверим статистическую значимость.
alpha = 0.05

# Рассчитаем суммы депозитов на пользователя для контрольной группы
control_deposits = (
    control_df[control_df['event_name'].isin(['first_deposit', 'second_deposit'])]
    .groupby('user_id')['amount']
    .sum()
)

# Рассчитаем суммы депозитов на пользователя для тестовой группы
test_deposits = (
    test_df[test_df['event_name'].isin(['first_deposit', 'second_deposit'])]
    .groupby('user_id')['amount']
    .sum()
)

# Проведем t-тест для независимых выборок (Welch)
stat_avg_deposit_per_user, p_key_avg_deposit_per_user = ttest_ind(
    control_deposits, 
    test_deposits, 
    equal_var=False,
    alternative='greater'
)

if p_key_avg_deposit_per_user >= alpha:
    print('Различия между средними суммам всех депозитов на одного пользователя в контрольной и тестовой группах статистически НЕ значимы.')
else:
    print('Различия между средними суммам всех депозитов на одного пользователя в тестовой группе статистически значимо больше.')

print(f'p_value = {p_key_avg_deposit_per_user.round(4)}')
````

Несмотря на то, что предыдущая вспомогательная метрика в открытие второго депозита показала рост количества вторых депозитов и абсолютные значения в тестовой (135.5) и контрольной (141.4) группах показали рост в тесте. Статистический тест показал, что разница между метриками статистически НЕ значима.

**Общий промежуточный вывод:**

````python
# Установим сид для воспроизводимости
np.random.seed(341)

# Рассчитаем суммы депозитов на пользователя для контрольной группы
control_deposits = (
    control_df[(control_df['event_name'].isin(['first_deposit', 'second_deposit']) & (control_df['amount'] > 0))]
    .groupby('user_id')['amount']
    .sum()
)

# Рассчитаем суммы депозитов на пользователя для тестовой группы
test_deposits = (
    test_df[(test_df['event_name'].isin(['first_deposit', 'second_deposit'])) & (test_df['amount'] > 0)]
    .groupby('user_id')['amount']
    .sum()
)

# установим количество бутстрап-итераций
n_iterations = 1000

# Перечислим перцентили для анализа
percentiles = [25, 50, 75]

results = {}

for p in percentiles:
    boot_diffs = []

    for _ in range(n_iterations):
        # Бутстрап выборки
        boot_control = np.random.choice(control_deposits, size=len(control_deposits), replace=True)
        boot_test = np.random.choice(test_deposits, size=len(test_deposits), replace=True)

        # Расчет выбранного перцентиля
        control_p = np.percentile(boot_control, p)
        test_p = np.percentile(boot_test, p)

        # Разница между тестовой и контрольной группой
        boot_diffs.append(test_p - control_p)

    # Преобразуем в массив
    boot_diffs = np.array(boot_diffs)
    
    # Доверительный интервал (95%) для разницы перцентилей
    ci_lower, ci_upper = np.percentile(boot_diffs, [2.5, 97.5])
    
    # Средние значения перцентилей без бутстрапа
    control_val = np.percentile(control_deposits, p)
    test_val = np.percentile(test_deposits, p)
    diff_mean = test_val - control_val
    
    results[p] = {
        'control_pct': control_val,
        'test_pct': test_val,
        'mean_diff': diff_mean,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'diff_array': boot_diffs
    }

# Вывод результатов
for p in percentiles:
    res = results[p]
    print(f"{p} перцентиль:")
    print(f"  Контроль: {res['control_pct']:.2f}")
    print(f"  Тест: {res['test_pct']:.2f}")
    print(f"  Разница (средняя): {res['mean_diff']:.2f}")
    print(f"  95% доверительный интервал: [{res['ci_lower']:.2f}, {res['ci_upper']:.2f}]\n")
````
